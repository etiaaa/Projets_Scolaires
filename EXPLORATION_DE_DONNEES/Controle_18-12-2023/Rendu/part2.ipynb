```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Charger les données depuis un fichier CSV
file_path = './duck_race_game_clean.csv'
data = pd.read_csv(file_path)

print(data.columns)
```

```python
# Afficher les valeurs de la colonne 'Standing'
print(data[' Standing'].unique())

# Créer une nouvelle colonne 'Total_Points' en utilisant la colonne ' Standing' comme référence
data['Total_Points'] = pd.to_numeric(data[' Standing'], errors='coerce')  # Convertir en numérique

# Filtrer les données pour les canards hors d'Europe (ajuste le nom de la colonne si nécessaire)
canards_hors_europe = data[data[' City'] != ' Europe']

# Regrouper par la ville et calculer le total des points
total_points_par_ville = canards_hors_europe.groupby(' City')['Total_Points'].sum()

# Imprimer les valeurs de total_points_par_ville
print(total_points_par_ville)

# Créer un graphique à barres
plt.figure(figsize=(10, 6))
total_points_par_ville.sort_values().plot(kind='barh', color='skyblue')
plt.title("Total des Points des Canards Hors d'Europe par Ville")
plt.xlabel('Total des Points')
plt.ylabel('Ville')
plt.show()
```
